```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
os.chdir(r'C:\Users\obuli\Karthi O\Statistics')
```

```python
file_airfare = pd.read_csv('airfares_v1.csv')
file_airfare [:10]
```

```python
file_airfare.isnull().sum()
```

### Distribution plots

#### displot
- Figure-level interface for drawing distribution plots onto a FacetGrid.

```python
#sns.set(rc={"figure.figsize": (8, 4)})
plt.figure(figsize=(30, 15))
sns.displot(file_airfare, x='DISTANCE', y='FARE', hue='SLOT')
plt.show()
```

```python
plt.figure(figsize=(30, 15))
sns.histplot(data=file_airfare, x="DISTANCE", y="FARE", hue='SLOT')
plt.xlabel('Distace')
plt.ylabel('Fare')
```

```python
plt.figure(figsize=(30, 15))
sns.kdeplot(data=file_airfare, x="DISTANCE", y="FARE", hue='SLOT')
plt.xlabel('Distace')
plt.ylabel('Fare')
```

```python
plt.figure(figsize=(30, 15))
sns.rugplot(data=file_airfare, x="PAX", y="FARE" )
plt.xlabel('Passengers')
plt.ylabel('Fare')
```

### Categorical plots

```python
plt.figure(figsize=(100, 40))
plt.title("Average Fare for the respective distance")
sns.barplot(x=file_airfare.DISTANCE, y=file_airfare.FARE)
plt.xlabel('Distace')
plt.ylabel('Fare')
```

```python
plt.figure(figsize=(100, 40))
plt.title("Average Fare for the respective distance")
sns.countplot(data=file_airfare)
```

### Matrix plot

```python
plt.figure(figsize=(30, 15))
sns.heatmap(data=pd.get_dummies(file_airfare['Departure_CITY']),annot=True, cmap='ocean')
##ValueError: could not convert string to float: 'Dallas/Fort Worth   TX', so trying with dummies
plt.xlabel('Departure City')
plt.ylabel('What is this?')
```

### Relational plots

```python
plt.figure(figsize=(30, 15))
plt.title("Distance Vs population")
sns.lineplot(data=file_airfare, x=file_airfare.DISTANCE, y=file_airfare.FARE, hue='COUPON')
```

```python
plt.figure(figsize=(10, 5))
sns.scatterplot(x=file_airfare['Departure_CITY'], y=file_airfare['FARE'], hue=file_airfare['GATE'])
```

```python
plt.figure(figsize=(10, 5))
sns.scatterplot(x=file_airfare['PAX'], y=file_airfare['FARE'], hue=file_airfare['COUPON'])
```

### Regression plot

```python
plt.figure(figsize=(30, 15))
sns.regplot(data=file_airfare, x="DISTANCE", y="FARE")
plt.xlabel('Distace')
plt.ylabel('Fare')
```

```python
plt.figure(figsize=(30, 15))
sns.regplot(data=file_airfare, x="Departure_avg_INCOME", y="Arrival_avg_INCOME")
```

```python
plt.figure(figsize=(30, 15))
sns.lmplot(data=file_airfare, x="DISTANCE", y="FARE", hue="GATE")
```

### Pair Grid

```python
plt.figure(figsize=(30, 15))
sns.pairplot(data=file_airfare, corner=True)
```

```python
plt.figure(figsize=(30, 15))
sns.PairGrid(data=file_airfare, x_vars=['Departure_CITY'], y_vars=['Arrival_CITY'])
```

```python
plt.figure(figsize=(30, 15))
sns.PairGrid(data=file_airfare, x_vars=['Departure_CITY'], y_vars=['FARE'])
```

### Joint Grid

```python
plt.figure(figsize=(30, 15))
sns.JointGrid(data=file_airfare, x='DISTANCE', y='FARE')
```

```python
plt.figure(figsize=(30, 15))
sns.jointplot(data=file_airfare, x='DISTANCE', y='FARE',hue='COUPON')
plt.title('jointplot sample')
sns.set_theme(color_codes='black')
```
